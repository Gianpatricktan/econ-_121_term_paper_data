```python
# Simple Linear Regression (Least Squares Point Estimates)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
```

```python
# x (independent variable) is the year
x = np.array([1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021])

# y (dependent variable) is the mean years of schooling
y = np.array([7.178850174,7.217683172,7.256516171,7.295349169,7.334182167,7.373015165,7.411848164,7.450681162,7.48951416,7.528347158,7.567180157,7.836127639,8.105075121,8.374022603,8.642970085,8.65989995,8.676829815,8.69375968,8.710689545,8.812839985,8.914990425,8.991986911,9.068983396,9.145979881,8.971109867,8.796239853,8.621369839,8.446499825,8.707799911,8.969099998,8.969099998,8.969099998])

# n is the sample size
n = np.size(x)


# x_mean is the same as x̄
x_mean = np.mean(x)

# y_mean is the same as ȳ
y_mean = np.mean(y)

x_mean,y_mean

# Formula of SSxy
SSxy = np.sum(x*y) - n*x_mean*y_mean

# Formula of SSxx
SSxx = np.sum(x*x) - n*x_mean*x_mean

# Formula of b1 (slope)
b1 = (SSxy / SSxx)

# Formula of b0 (y-intercept)
b0 = y_mean - b1*(x_mean)

# Scatter Plot
plt.scatter(x,y)
plt.xlabel('Year') # Independent Variable: x
plt.ylabel('Mean Years of Schooling') # Dependent Variable: y
```

```python
# Prediction of y (same as ŷ)
y_pred = b0 + b1*(x)


# Creating the scatter plot and the regression line
plt.title('Mean Years of Schooling (Philippines)')
plt.scatter(x, y, color = 'blue') # Actual data
plt.plot(x, y_pred, color = 'red') # Predicted data
plt.xlabel('Year') # Independent Variable: x
plt.ylabel('Mean Years of Schooling') # Dependent Variable: y
plt.ylim(bottom = 0, top = 10) # Set the range of values of the y-axis

print(f'''Equation of Simple Linear Regression Model: 
ŷ = {b1}x + {b0}''')
```
